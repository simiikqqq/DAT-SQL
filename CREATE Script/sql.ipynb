### 1. Tabulka product obsahuje atribut description. Jaká data se v něm nacházejí a jakou mají strukturu?

#### Sloupec description

Sloupec **description** v tabulce `product` je typu **JSON**.

Podle vložených dat může obsahovat například:
- brand – značka produktu
- category – kategorie produktu
- cpu – typ procesoru

Každý produkt může mít jinou strukturu JSON podle typu produktu.

### Příklad hodnoty

{
    "brand":"Dell",
    "category":"laptop",
    "cpu":"Intel i5",
    "ram_gb":8
}

nebo
{
"category":"cable",
"type":"Cat6",
"length_m":2
}

### 2. Co provede dotaz:
SELECT
name,
JSON_VALUE(description, '$.category') AS category
FROM product; 

```python
SELECT
name, -- název produktu
JSON_VALUE(description, '$.category') AS category -- přečte z JSON sloupce description hodnotu category
FROM product; -- z tabulky product
```

### 3. Vytvoř SQL pro zobrazení všech produktů z vybraného skladu.

```python
SELECT *
FROM product
WHERE warehouse_id = 1; -- vybere produkty uložené ve skladu s id 1
```

### 4. Vytvoř SQL pro zobrazení všech produktů, které mají ve sloupci description v category uvedeno cable.

```python
SELECT *
FROM product
WHERE JSON_VALUE(description, '$.category') = 'cable'; -- vybere produkty, kde je v JSON category = cable
```

### 5. Vytvoř SQL pro zobrazení všech produktů, jejichž počet je větší než 20 seřezený vzestupně podle počtu.

```python
SELECT *
FROM product
WHERE quantity > 20 -- vybere produkty s množstvím větším než 20
ORDER BY quantity ASC; -- seřadí výsledky podle quantity vzestupně
```

### 6. Vytvoř SQL pro zobrazení všech produktů, které byly uskladněny v březnu 2024.

```python
SELECT *
FROM product
WHERE stored_from BETWEEN '2024-03-01' AND '2024-03-31'; -- vybere produkty uložené v březnu 2024
```
